```python
from rake_nltk import Rake
import pandas as pd
import numpy as np
import os
```

```python
r = Rake(punctuations= ['(', ')', ',', ':', '),', ').', '.'])
```

```python
os.chdir('/Users/alisdghnia/Desktop/')
df2 = pd.read_hdf('all text.h5')
df2
```

```python
corpus_data = pd.DataFrame({'corpus': [''.join(df2['test'].tolist())]})
```

```python
corpus_data
```

```python
df2.reset_index(inplace=True)
df2.rename(columns= {'index' : 'Name'}, inplace=True)

df2
```

```python
r.extract_keywords_from_text(corpus_data.corpus[0])
```

```python
df = pd.DataFrame(r.get_ranked_phrases_with_scores(), columns= ['score', 'phrase'])
```

```python
df
```

```python
df = pd.DataFrame.from_dict(r.get_word_frequency_distribution(), orient='index')
```

```python
df.rename(columns={'0' : 'count'}, inplace=True)
```

```python
df.reset_index(inplace=True)
df.rename(columns= {'index' : 'word'}, inplace=True)
```

```python
df
```

```python
df.rename(columns={0 : 'score'}, inplace=True)
```

```python
df
```

```python
df_new = df.loc[df['score'] < 200000]
```

```python
df_new_new = df_new.loc[df_new['score'] > 100]
```

```python
df_new_new
```

```python
final_df = df_new_new[df_new_new['word'].str.len() > 2]
```

```python
final_df.reset_index(inplace=True)
```

```python
final_df
```

```python
final_df.drop(columns= ['index'], inplace=True)
```

```python
final_df
```

```python
final_df.to_csv('keywords_rake_library.csv')
```
