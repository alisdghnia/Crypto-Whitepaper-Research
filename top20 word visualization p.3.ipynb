```python
df_findings = []
for i in term_list:
    top = all_df[i].quantile(0.98)
    top_df = all_df[all_df[i] > top]
    count = len(top_df[np.log10(top_df.market_cap+1) != 0])
    sec_count = len(top_df[np.log10(top_df.market_cap+1) == 0])
    ratio = count / sec_count
    average = np.average(top_df[i])
    df_findings.append([i, count, sec_count, round(ratio, 2), round(average, 6), round(np.average(top_df.market_cap), 3)])
```
