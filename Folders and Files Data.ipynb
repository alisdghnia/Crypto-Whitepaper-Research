```python
import os
import fnmatch
import os
import sys
import re
import time
import PyPDF2
import xlsxwriter
row1 = 0
col1 = 0
row = 0
col = 0
workbook = xlsxwriter.Workbook('Users/alisdghnia/Desktop/wordcount.xlsx')
worksheet = workbook.add_worksheet()

path ="/Users/alisdghnia/Desktop/symbol"
#we shall store all the file names in this list
filelist = []

for root, dirs, files in os.walk(path):
	for file in files:
        #append the file name to the list
		filelist.append(os.path.join(root,file))

#print all the file names
#for name in filelist:
#    print(name)
filelistName = []
for root, dirs, files in os.walk(path):
    for fileName in files:
        filelistName.append(fileName)

filteredName = fnmatch.filter(filelistName, '*.pdf')
##Filelistname to filelist for the paths

for xx in filteredName:
##not filteredname but filelist instead
    print (xx)
    #worksheet.write(row, col, xx)
    row += 1
filtered = fnmatch.filter(filelist, '*.pdf')
print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")

for name in filtered:
    print(name)

print("end of script")

for pdf_file in filtered:
    def getPageCount(pdf_file):
        pdfFileObj = open(pdf_file, 'rb')
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
        pages = pdfReader.numPages
        return pages


    def extractData(pdf_file, page):
        pdfFileObj = open(pdf_file, 'rb')
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
        pageObj = pdfReader.getPage(page)
        data = pageObj.extractText()
        return data


    def getWordCount(data):
        data = data.split()
        return len(data)



    # get the word count in the pdf file
    totalWords = 0
    numPages = getPageCount(pdf_file)
    for i in range(numPages):
        text = extractData(pdf_file, i)
        totalWords += getWordCount(text)
    time.sleep(1)

    print(totalWords)
    worksheet.write(row1, col1+1, totalWords)
    row1 += 1
# fileNames = []
# for files in os.walk(path):
# 	for file in files:
#         #append the file name to the list
# 		fileNames.append(file)

# filteredName = fnmatch.filter(fileNames,'*.pdf')




###workbook.close()
```
