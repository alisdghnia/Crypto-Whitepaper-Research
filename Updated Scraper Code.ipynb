```python
from seleniumwire import webdriver
from selenium.webdriver.chrome.options import Options

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.support.ui import Select

import time
from selenium.webdriver.common.keys import Keys

# data manipulate library
import pandas as pd

import csv

import stopit
```

```python
# Setup selenium

chrome_options = Options()

# chrome_options.add_argument('disable-popup-blocking')
# chrome_options.headless = True
# path_to_extension = r'C:\Users\minht\Desktop\5.3.0_0'

# chrome_options.add_argument('load-extension='+path_to_extension)

driver = webdriver.Chrome(executable_path='chromedriver.exe',chrome_options=chrome_options)
```

```python
path_to_result = 'result.csv'
```

```python
# read from csv
df = pd.read_csv('New-Coin-List.csv')
df
```

```python
# Write the header
fieldnames = ['coin_name','pdf_link']

with open(path_to_result, 'w', encoding='UTF8', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
```

```python
def WriteResultCsv(result):
    with open(path_to_result, 'a', encoding='UTF8', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writerow(result)
```

```python
def GetPdfLink(middle_link):
    pdf_link = ''

    del driver.requests
    
    try:
        with stopit.ThreadingTimeout(50, swallow_exc=True) as context_manager:

            driver.get(middle_link)

            paper = WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.XPATH, '//canvas'))
            )   

            # sample code we want to run...
            for request in driver.requests:
                if  (request.response):
                    if request.url.endswith(".pdf"):
                        pdf_link = str(request.url)
                        break

        if context_manager.state == context_manager.EXECUTED:
            return pdf_link   


        # Did code timeout?
        elif context_manager.state == context_manager.TIMED_OUT:
            pdf_link = 'DID NOT FINISH'
            return pdf_link


    except:
        pdf_link = 'COULD NOT FIND'
        return pdf_link
```

```python
for index, row in df.iterrows():
    # print(row['coin_name'], row['middle_link'])
    coin_name = row['coin_name']
    middle_link = row['middle_link']

    pdf_link = GetPdfLink(middle_link=middle_link)
    print(coin_name,pdf_link)
    result = {'coin_name': coin_name, 'pdf_link':pdf_link}
    WriteResultCsv(result=result)

    
```
