```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
os.chdir('/Users/alisdghnia/Desktop/Research Datasets/')
```

```python
df = pd.read_csv('names-mc-volume.csv')
df_1 = pd.read_csv('chat-gpt-top-20-tfidf.csv')
```

```python
all_df = pd.merge(df, df_1, left_on = 'Name', right_on = 'Unnamed: 0')
all_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_values = scaler.fit_transform(all_df.iloc[:, 5:])
scaled_values
```

```python
all_df.iloc[:, 5:] = scaled_values
all_df
```

```python
all_df = all_df.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y'])
```

```python
all_df
```

```python
top_5 = all_df['cryptocurrency'].quantile(0.95)
top_df = all_df[all_df['cryptocurrency'] > top_5]
top_df
```

```python
plt.scatter(x = np.log(all_df.market_cap), y=np.log(all_df.volume_y), s=all_df.cryptocurrency * 10, alpha=0.5)
plt.show()
```
